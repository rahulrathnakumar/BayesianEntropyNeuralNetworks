```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
def kl_divergence(p, q):
    '''
    Parameters:
    p: np.array of shape (n, )
    q: np.array of shape (n, )
    This function returns the Kullback-Leibler divergence between p and q.
    '''
    eps = 1e-9
    p = p + eps
    q = q + eps
    return np.sum(q * np.log(q/p))

def gaussian_log_likelihood(y_gt, mu, sigma):
    '''
    Parameters:
    '''
    return 1/2 * np.log(sigma**2) + 1/2 * (y_gt - mu)**2 / sigma**2
```

```python
x = np.arange(-10,10, 0.001)

# p is a uniform distribution with (a,b) = (-1,1)
p = stats.uniform.pdf(x, -20, 40)

q = stats.norm.pdf(x, 0, 12.65)

plt.plot(x, p, label='p(x)')
plt.plot(x, q, label='q(x)')

plt.legend()

kl = kl_divergence(p, q)
print('KL divergence:', kl)
```
